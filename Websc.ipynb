```python
title = Soup2.find(id='productTitle').get_text()
print(title)
```

```python
#Breaking the price into 3 parts, extracting each one individually and recombining them.


price_symbol = Soup2.find(class_='a-price-symbol').get_text(strip=True)
price_whole = Soup2.find(class_='a-price-whole').get_text(strip=True)
price_fraction = Soup2.find(class_='a-price-fraction').get_text(strip=True)
price = f'{price_symbol}{price_whole}{price_fraction}'
print(price)
```

```python
#number of reviews 

reviews_tag = Soup2.select_one("#acrCustomerReviewText")
if reviews_tag:
    reviews_text = reviews_tag.get_text(strip=True)        # "3,214 ratings"
    reviews_count = "".join(ch for ch in reviews_text.split()[0] if ch.isdigit())
    print("Reviews:", reviews_count)
```

```python

rating_text = Soup2.select_one("span.a-icon-alt").get_text(strip=True)
print(rating_text)         # "5.0 out of 5 stars"

rating_value = rating_text.split()[0]
print(rating_value)        # "5.0"
```

```python
#cleaning price and title for representation
title = title.strip()
print(title)
print(price)
```
